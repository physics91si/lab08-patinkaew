**Part 1: Theme Park Analysis**

```python
import numpy as np
from scipy import integrate
import matplotlib.pyplot as plt
```

```python
t_stamp, vel = np.loadtxt('droptower_vdata.txt', unpack = True)
pos = 30 + integrate.cumtrapz(vel, t_stamp, initial=0)
acc = np.diff(vel)
f, (sp0, sp1, sp2) = plt.subplots(1,3)
sp0.set_title('Position')
sp0.plot(t_stamp, pos, 'c')
sp1.set_title('Velocity')
sp1.plot(t_stamp, vel, color='purple')
sp2.set_title('Acceleration')
sp2.plot(t_stamp[:len(t_stamp)-1], acc, color='red')
plt.tight_layout()
plt.suptitle('Data Analysis of Drop Tower')
plt.subplots_adjust(top=0.85)
plt.show()
```

**Part 2: Drop Tower Continued**

```python
positive_acc = acc[acc > 0]
mean = np.mean(positive_acc)
print('average of positive acceleration is %.2f m/s^2' %mean)
```
